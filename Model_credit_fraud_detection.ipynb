Para entender mejor la naturaleza de cada variable en el dataset, se generan histogramas individuales para todas las columnas. Esto permite visualizar la distribuci√≥n de los datos, detectar sesgos, outliers y patrones relevantes.

El siguiente c√≥digo define una funci√≥n `draw_histograms()` que:

```python
def plot_time_density(dataframe, time_col='Time', class_col='Class'):
    """
    Genera un gr√°fico de densidad para comparar la distribuci√≥n del tiempo
    entre transacciones leg√≠timas y fraudulentas.

    Par√°metros:
    ----------
    dataframe : pd.DataFrame
        Dataset que contiene las columnas de tiempo y clase.
    time_col : str
        Nombre de la columna que representa el tiempo.
    class_col : str
        Nombre de la columna que representa la clase (0 = No Fraude, 1 = Fraude).
    """
    # Separar las clases
    class_0 = dataframe.loc[dataframe[class_col] == 0, time_col]
    class_1 = dataframe.loc[dataframe[class_col] == 1, time_col]

    # Crear el gr√°fico
    plt.figure(figsize=(12, 6))
    sns.kdeplot(class_0, label='No Fraude', color='midnightblue', linewidth=2)
    sns.kdeplot(class_1, label='Fraude', color='darkred', linewidth=2)

    # Personalizaci√≥n
    plt.title('Distribuci√≥n de Tiempo en Transacciones (Densidad)', fontsize=14, color='DarkRed')
    plt.xlabel('Tiempo [segundos]', fontsize=12)
    plt.ylabel('Densidad', fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Llamada a la funci√≥n
plot_time_density(creditcard)
```

## ‚úÖ Conclusi√≥n sobre la distribuci√≥n del tiempo
Las transacciones fraudulentas presentan una distribuci√≥n m√°s uniforme a lo largo del tiempo en comparaci√≥n con las transacciones leg√≠timas. Esto significa que los fraudes ocurren en diferentes momentos del d√≠a, incluyendo periodos de baja actividad real, como la noche.
Por el contrario, las transacciones v√°lidas tienden a concentrarse en ciertos intervalos, reflejando patrones normales de uso.
Implicaci√≥n para el modelo:
La variable Tiempo puede ser √∫til para la detecci√≥n de fraude, especialmente si se transforma en caracter√≠sticas como rangos horarios o indicadores de actividad nocturna, ya que los fraudes no siguen el mismo patr√≥n temporal que las transacciones leg√≠timas.

```python
def plot_amount_boxplots(dataframe, amount_col='Amount', class_col='Class'):
    """
    Genera dos boxplots para comparar la distribuci√≥n del monto entre clases:
    - Uno mostrando outliers.
    - Otro ocultando outliers.
    """
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

    # Boxplot con outliers
    sns.boxplot(ax=ax1, x=class_col, y=amount_col, hue=class_col,
                data=dataframe, palette="PRGn", showfliers=True)
    ax1.set_title("Montos por Clase (con outliers)", color='DarkRed')

    # Boxplot sin outliers
    sns.boxplot(ax=ax2, x=class_col, y=amount_col, hue=class_col,
                data=dataframe, palette="PRGn", showfliers=False)
    ax2.set_title("Montos por Clase (sin outliers)", color='DarkRed')

    plt.tight_layout()
    plt.show()

# Llamada a la funci√≥n
plot_amount_boxplots(creditcard)
```

En esta secci√≥n se analizar√°n las variables del dataset para identificar patrones que puedan ser √∫tiles en la detecci√≥n de fraude. Se comparar√° la distribuci√≥n de cada caracter√≠stica entre transacciones leg√≠timas (Class = 0) y fraudulentas (Class = 1) mediante gr√°ficos de densidad.


```python
def plot_feature_distributions(dataframe, features, class_col='Class'):
    """
    Genera gr√°ficos de densidad para cada feature, comparando las clases 0 y 1.

    """
    t0 = dataframe.loc[dataframe[class_col] == 0]
    t1 = dataframe.loc[dataframe[class_col] == 1]

    sns.set_style('whitegrid')
    fig, ax = plt.subplots(8, 4, figsize=(16, 28))  # Ajustar seg√∫n cantidad de features
    ax = ax.flatten()

    for i, feature in enumerate(features):
        sns.kdeplot(t0[feature], bw_adjust=0.5, label="No Fraude", ax=ax[i], color='midnightblue')
        sns.kdeplot(t1[feature], bw_adjust=0.5, label="Fraude", ax=ax[i], color='darkred')
        ax[i].set_xlabel(feature, fontsize=12)
        ax[i].tick_params(axis='both', which='major', labelsize=10)
        ax[i].legend()

    plt.tight_layout()
    plt.show()

# Llamada a la funci√≥n
var = creditcard.columns.drop(['Class'])  # Excluye la variable objetivo
plot_feature_distributions(creditcard, var)
```

## üöÄ Inicio del proceso de modelado

En esta secci√≥n comenzaremos a construir y evaluar modelos predictivos para la detecci√≥n de fraude en transacciones con tarjeta de cr√©dito.  
El objetivo es aplicar algoritmos adecuados, considerando el fuerte desbalance de clases, y utilizar m√©tricas apropiadas como **Precision**, **Recall**, **F1-score** y **AUC-ROC** para evaluar el rendimiento.  
Tambi√©n se explorar√°n t√©cnicas de balanceo y ajustes de hiperpar√°metros para optimizar los resultados.

```python
 target = 'Class'
predictors = ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',\
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',\
       'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28',\
       'Amount']
```

Separamos la data en los sets de Validaci√≥n, Test y Train:

```python
from sklearn.model_selection import train_test_split

# Par√°metros de divisi√≥n
TEST_SIZE = 0.2      # 20% para prueba
VALID_SIZE = 0.2     # 20% del conjunto de entrenamiento para validaci√≥n
RANDOM_STATE = 42    # Semilla para reproducibilidad

# Divisi√≥n inicial: Train + Test
train_df, test_df = train_test_split(creditcard, test_size=TEST_SIZE,
                                      random_state=RANDOM_STATE, shuffle=True)

# Divisi√≥n secundaria: Train + Valid
train_df, valid_df = train_test_split(train_df, test_size=VALID_SIZE,
                                      random_state=RANDOM_STATE, shuffle=True)

print(f"‚úÖ Train: {train_df.shape}")
print(f"‚úÖ Valid: {valid_df.shape}")
print(f"‚úÖ Test: {test_df.shape}")
```

## üå≤ Entrenamiento del modelo: RandomForestClassifier

Comenzamos el proceso de modelado utilizando **RandomForestClassifier**, un algoritmo basado en ensambles de √°rboles que maneja bien datos desbalanceados y relaciones no lineales.

**Par√°metros iniciales:**
- `n_estimators = 100` (n√∫mero de √°rboles)
- `criterion = 'gini'` (criterio de divisi√≥n)
- `n_jobs = 4` (procesamiento paralelo)
- `random_state = 42` (reproducibilidad)

**M√©trica de validaci√≥n:**  
Se usar√° **GINI**, calculado como:
\[
GINI = 2 \times AUC - 1
\]
donde AUC es el √°rea bajo la curva ROC.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# Par√°metros del modelo
NO_JOBS = 4
RANDOM_STATE = 42
RFC_METRIC = 'gini'
NUM_ESTIMATORS = 100

# Inicializaci√≥n del modelo
clf = RandomForestClassifier(n_jobs=NO_JOBS,
                             random_state=RANDOM_STATE,
                             criterion=RFC_METRIC,
                             n_estimators=NUM_ESTIMATORS,
                             verbose=False)

# Entrenamiento
clf.fit(train_df[predictors], train_df[target].values)

# Predicciones en validaci√≥n
preds = clf.predict(valid_df[predictors])

# Importancia de caracter√≠sticas
tmp = pd.DataFrame({'Feature': predictors, 'Feature importance': clf.feature_importances_})
tmp = tmp.sort_values(by='Feature importance', ascending=False)

plt.figure(figsize=(7, 4))
plt.title('Importancia de las caracter√≠sticas', fontsize=14)
s = sns.barplot(x='Feature', y='Feature importance', data=tmp)
s.set_xticklabels(s.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()

# Confusion Matrix
cm = pd.crosstab(valid_df[target].values, preds, rownames=['Actual'], colnames=['Predicted'])
fig, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(cm,
            xticklabels=['No Fraude', 'Fraude'],
            yticklabels=['No Fraude', 'Fraude'],
            annot=True, linewidths=.2, linecolor="Darkblue", cmap="Blues", ax=ax)
plt.title('Matriz de Confusi√≥n', fontsize=14)
plt.show()

# ROC-AUC
roc_score = roc_auc_score(valid_df[target].values, preds)
print(f"‚úÖ ROC-AUC Score: {roc_score:.3f}")
```

## ‚ö° Entrenamiento del modelo: XGBoost

XGBoost es un algoritmo basado en **Gradient Boosting**, muy eficiente para problemas de clasificaci√≥n con datos desbalanceados.

**Par√°metros iniciales:**
- `objective = 'binary:logistic'` (clasificaci√≥n binaria)
- `eta = 0.039` (learning rate)
- `max_depth = 2` (profundidad m√°xima de los √°rboles)
- `subsample = 0.8` (submuestreo de filas)
- `colsample_bytree = 0.9` (submuestreo de columnas)
- `eval_metric = 'auc'` (m√©trica de evaluaci√≥n)
- `random_state = 42` (reproducibilidad)

Se utilizar√° **early stopping** para evitar sobreajuste y se monitorizar√°n las m√©tricas en los conjuntos de entrenamiento y validaci√≥n.

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score

# Preparar datasets para XGBoost
dtrain = xgb.DMatrix(train_df[predictors], train_df[target].values)
dvalid = xgb.DMatrix(valid_df[predictors], valid_df[target].values)
dtest = xgb.DMatrix(test_df[predictors], test_df[target].values)

# Watchlist para monitoreo
watchlist = [(dtrain, 'train'), (dvalid, 'valid')]

# Par√°metros del modelo
params = {
    'objective': 'binary:logistic',
    'eta': 0.039,
    'max_depth': 2,
    'subsample': 0.8,
    'colsample_bytree': 0.9,
    'eval_metric': 'auc',
    'random_state': RANDOM_STATE
}

# Entrenamiento con early stopping
MAX_ROUNDS = 1000
EARLY_STOP = 50
VERBOSE_EVAL = 50

model = xgb.train(params,
                  dtrain,
                  MAX_ROUNDS,
                  watchlist,
                  early_stopping_rounds=EARLY_STOP,
                  maximize=True,
                  verbose_eval=VERBOSE_EVAL)

# Importancia de caracter√≠sticas
fig, ax = plt.subplots(figsize=(8, 5))
xgb.plot_importance(model, height=0.8, title="Importancia de caracter√≠sticas (XGBoost)", ax=ax, color="green")
plt.tight_layout()
plt.show()

# Predicciones en test
preds = model.predict(dtest)

# ROC-AUC en test
roc_score = roc_auc_score(test_df[target].values, preds)
print(f"‚úÖ ROC-AUC Score en test: {roc_score:.3f}")
```
